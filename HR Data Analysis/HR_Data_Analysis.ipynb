#HR Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Loading Dataset

```python
df = pd.read_csv('HR Data.csv')
print(df.head())
print(df.info())
```

#Data Preprocessing and Data Cleaning

```python
print("Missing Values in the Dataset")
print(df.isnull().sum())
```

```python
df['DailyRate'] = df['DailyRate'].astype(int)
mean_daily_rate = df['DailyRate'].mean()
df.fillna({'DailyRate':mean_daily_rate }, inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18'], inplace=True)
```

#Eliminating the NaN values

```python
df.dropna()
df.head()
```

#Eliminating redundant entries

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df['Gender'] = df['Gender'].str.capitalize()
```

```python
df.head()
```

# Renaming Columns

```python
df.rename(columns={
    'BusinessTravel': 'Business_Travel',
    'DailyRate': 'Daily_Rate',
    'Department': 'Department',
    'DistanceFromHome': 'Distance',
    'EducationField': 'Education_Field',
    'EmployeeNumber': 'Employee_Number',
    'EnvironmentSatisfaction': 'Environmental_Satisfaction',
    'HourlyRate': 'Hourly_Rate',
    'JobInvolvement': 'Job_Involvement',
    'JobLevel': 'Job_Level',
    'JobRole': 'Job_Role',
    'JobSatisfaction': 'Job_Satisfaction',
    'MaritalStatus': 'Marital_Status',
    'MonthlyIncome': 'Monthly_Income',
    'MonthlyRate': 'Monthly_Rate',
    'NumCompaniesWorked': 'Companies_Worked_Count',
    'OverTime': 'Over_Time',
    'PercentSalaryHike': 'Percent_Salary_Hike',
    'PerformanceRating': 'Performance_Rating',
    'RelationshipSatisfaction': 'Relationship_Satisfaction',
    'StockOptionLevel': 'Stock_Option_Level',
    'TotalWorkingYears': 'Total_Working_Years',
    'TrainingTimesLastYear': 'Training_Times_Last_Year',
    'WorkLifeBalance': 'Work_Life_Balance',
    'YearsAtCompany': 'Years_At_Company',
    'YearsInCurrentRole': 'Years_In_Current_Role',
    'YearsSinceLastPromotion': 'Years_Since_Last_Promotion',
    'YearsWithCurrManager': 'Years_With_Current_Manager'
}, inplace=True)
df.head()
```

#EDA And Visualization

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department')
plt.title('Employees per Department')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Gender', y='MonthlyIncome')
plt.title('Salary by Gender')
plt.show()
```

```python
sns.histplot(data = df,x='Age',kde =True,hue='Attrition', palette={'Yes': 'pink', 'No': 'green'})
```

```python
attrition_by_role = df.groupby(['JobRole', 'Attrition']).size().unstack(fill_value=0)
attrition_by_role.plot(kind='bar', stacked=True, figsize=(10,5), color=['blue', 'pink'])
plt.title("Attrition by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['MonthlyIncome'],kde=True,bins=25,color='red')
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Department', y='JobSatisfaction',color='yellow')
plt.title('Average Job Satisfaction per Department')
plt.xticks(rotation=15)
plt.show()
```

```python
job_roles = df['JobRole'].value_counts()
plt.figure(figsize=(10,5))
plt.barh(job_roles.index, job_roles.values, color='skyblue')
plt.title("Job Role")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.tight_layout()
plt.show()
```

```python
from numpy import triu_indices_from
import numpy as np

# Selecting numeric attrition-related features
attrition_factors = df[['Age', 'JobSatisfaction', 'WorkLifeBalance', 'YearsAtCompany', 'Attrition']].copy()
attrition_factors['Attrition'] = attrition_factors['Attrition'].map({'Yes': 1, 'No': 0})

corr = attrition_factors.corr()

fig, ax = plt.subplots(figsize=(10, 6))
cax = ax.matshow(corr, cmap='Blues')
plt.title('Correlation Matrix - Attrition Factors', pad=20)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
fig.colorbar(cax)
plt.tight_layout()
plt.show()
```

```python
demo = df.groupby(['Gender', 'MaritalStatus']).size().unstack(fill_value=0)
demo.plot(kind='bar', stacked=True, figsize=(8,5), colormap='tab20')
plt.title("Employee Demographics by Gender and Marital Status")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
